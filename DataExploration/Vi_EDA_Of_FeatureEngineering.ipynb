```python
# Defining colors for each Bin
colors = ['yellow', 'orchid', 'turquoise', 'mediumspringgreen', 'cornflowerblue', 'indianred']
bin_labels = ['0 to 30 Days', '31 to 90 Days', '91 to 180 Days', '181 to 270 Days', '271 to 360 Days', '360+ Days']
```

```python
longevity_0_daily_report
```

```python
# Calculate total deposit by each bin
total_deposits = [
    longevity_0_daily_report['net_deposit'].sum(),
    longevity_1_daily_report['net_deposit'].sum(),
    longevity_2_daily_report['net_deposit'].sum(),
    longevity_3_daily_report['net_deposit'].sum(),
    longevity_4_daily_report['net_deposit'].sum(),
    longevity_5_daily_report['net_deposit'].sum()
]

import matplotlib.pyplot as plt

# Creating the plot
plt.figure(figsize=(16, 9))

# Plotting the bar graph
bars = plt.bar(range(len(total_deposits)), total_deposits, tick_label=bin_labels, color=colors)

# Adding labels for the totals on top of each bar
for bar, total in zip(bars, total_deposits):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 5000, f'{total:,.2f}', ha='center', va='bottom', fontsize=16, color='black')

# Adding labels and title
plt.xlabel('Longevity Bins', fontsize=18)
plt.ylabel('Total Deposit', fontsize=18)
plt.title('Total Deposit by Longevity Bin', fontsize=22)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Displaying the graph
plt.tight_layout()
plt.show()
```
