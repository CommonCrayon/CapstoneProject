```python
# Read the CSV file into a DataFrame
login_dataframe = pd.read_csv("Dataset/login.csv")

# Remap the 'country' column using the country_mapping dictionary
login_dataframe['country'] = login_dataframe['country'].map(country_mapping).fillna(login_dataframe['country'])

# Remove the 'account_currency' and 'reg_date' columns
login_dataframe.drop(columns=['account_currency'], inplace=True)

login_dataframe.head()
```

```python
# Read the CSV file into DataFrames
trades_dataframe = pd.read_csv("Dataset/trades.csv")

# Grouping by 'login' and counting the total number of tickets
total_tickets_per_login = trades_dataframe.groupby('login')['ticket'].count()

# Filtering rows where 'cmd' is 1 and counting the number of such tickets per login
cmd_1_tickets_per_login = trades_dataframe[trades_dataframe['cmd'] == 1].groupby('login')['ticket'].count()

# Calculating the percentage of tickets where 'cmd' is 1
percentage_cmd_1_per_login = (cmd_1_tickets_per_login / total_tickets_per_login) * 100

# Grouping by 'login' and calculating the average volume of trades per login
average_volume_per_login = trades_dataframe.groupby('login')['volume'].mean()
average_volume_usd_per_login = trades_dataframe.groupby('login')['volume_usd'].mean()

average_open_price_per_login = trades_dataframe.groupby('login')['open_price'].mean()
average_close_price_per_login = trades_dataframe.groupby('login')['close_price'].mean()


# Filter out zero values for TP and SL
non_zero_tp_dataframe = trades_dataframe[trades_dataframe['tp'] != 0]
non_zero_sl_dataframe = trades_dataframe[trades_dataframe['sl'] != 0]

# Compute the average TP and SL per login
average_tp_per_login = non_zero_tp_dataframe.groupby('login')['tp'].mean()
average_sl_per_login = non_zero_sl_dataframe.groupby('login')['sl'].mean()

average_commission_per_login = trades_dataframe.groupby('login')['commission'].mean()
average_swaps_per_login = trades_dataframe.groupby('login')['swaps'].mean()
average_profit_per_login = trades_dataframe.groupby('login')['profit'].mean()

reason_per_login = trades_dataframe.groupby('login')['reason'].apply(lambda x: x.value_counts().idxmax())

# Creating a DataFrame from the series
result_dataframe = pd.DataFrame({
    'Total_Trades' : total_tickets_per_login,
    'Sell_Percentage': percentage_cmd_1_per_login, 
    'Average_volume': average_volume_per_login,
    'Average_volume_usd': average_volume_usd_per_login,
    'Average_Open_Price': average_open_price_per_login,
    'Average_Close_Price': average_close_price_per_login,
    'Average_take_profit': average_tp_per_login,
    'Average_stop_loss': average_sl_per_login,
    'Trading_Method' : reason_per_login,
    'Average_Commision' : average_commission_per_login,
    'Average_Swaps' : average_swaps_per_login,
    'Average_Profit' : average_profit_per_login,
    })
```

```python
# Merge the two dataframes on the 'login' column
merged_dataframe = pd.merge(login_dataframe, result_dataframe, on='login', how='inner')

# Display the merged dataframe
merged_dataframe.head()
```
